## Q1

Find the Jeffreys prior for $\theta$ where $y \mid \theta \sim \text{Bernoulli}(\sin^2\theta)$.

First, the Fisher information:
$$
\begin{align}
I(\theta) &= - E_{y \mid \theta}\left[ \frac{\partial^2}{\partial \theta^2} \log p(y \mid \theta) \right]  \\
&= - E_{y \mid \theta}\left[ \frac{\partial^2}{\partial \theta^2} \log \left[(\sin^2(\theta))^y (1 - \sin^2(\theta))^{1-y} \right]  \right] \\
&= - E_{y \mid \theta}\left[ \frac{\partial^2}{\partial \theta^2} \log \left[(\sin^2(\theta))^y (\cos^2(\theta))^{1-y} \right]  \right] && \text{Using } sin^2(\theta) + cos^2(\theta) = 1\\
&= - E_{y \mid \theta}\left[ \frac{\partial^2}{\partial \theta^2} \log \left[(\sin(\theta))^{2y} (\cos(\theta))^{2(1-y)} \right]  \right] \\
&= - E_{y \mid \theta}\left[ \frac{\partial^2}{\partial \theta^2} 2y \log (\sin(\theta)) + 2(1-y)\log(\cos(\theta))  \right] \\
&= - E_{y \mid \theta}\left[ \frac{\partial}{\partial \theta} 2y \frac{\cos(\theta)}{\sin(\theta)} - 2(1-y)\frac{\sin(\theta)}{\cos(\theta)}  \right] \\
&= - E_{y \mid \theta}\left[ 2y \left(\frac{-\sin^2(\theta)-\cos^2(\theta)}{\sin^2(\theta)}\right) - 2(1-y)\left(\frac{\cos^2(\theta)+\sin^2(\theta)}{\cos^2(\theta)}\right)  \right] \\
&= - E_{y \mid \theta}\left[ 2y \left(\frac{-1}{\sin^2(\theta)}\right) - 2(1-y)\left(\frac{1}{\cos^2(\theta)}\right)  \right] && \text{Using } sin^2(\theta) + cos^2(\theta) = 1 \\
&= - E_{y \mid \theta}\left[ -\frac{2y}{\sin^2(\theta)} - \frac{2(1-y)}{\cos^2(\theta)}  \right] \\
&= - \left[ -\frac{2E(y \mid \theta)}{\sin^2(\theta)} - \frac{2(1-E(y \mid \theta))}{\cos^2(\theta)}  \right] \\
&= - \left[ -\frac{2\sin^2(\theta)}{\sin^2(\theta)} - \frac{2(1-\sin^2(\theta))}{\cos^2(\theta)}  \right] \\
&= - \left[ -2 - 2  \right] &&\text{Again, using } sin^2(\theta) + cos^2(\theta) = 1\\
&= 4\\
\end{align}
$$
Therefore, Jeffreys prior $p(\theta) \propto \sqrt{I(\theta)} \propto 2 \propto 1$, also known as a flat prior.

## Q2

Likelihood: $y_i \mid \mu \sim^{iid} N(\mu, 1) \space \text{for} \space i=1, ..., n$.

Prior: $\mu \sim .5N(-1, 1) + .5N(1, 1)$.

Single observation: $y=1$.

$$p(\mu) = \frac{1}{\sqrt{2\pi}} \left( 0.5 e^{-\frac{1}{2}(\mu + 1)^2} + 0.5 e^{-\frac{1}{2}(\mu - 1)^2} \right)$$

Hints:

$$\phi(x; \mu_1, \sigma_1^2)\phi(x; \mu_2, \sigma_2^2) = \phi\left(x; \frac{\mu_1/\sigma_1^2 + \mu_2/\sigma_2^2}{1/\sigma_1^2 + 1/\sigma_2^2}, \frac{1}{1/\sigma_1^2 + 1/\sigma_2^2}\right)\phi\left(\mu_1 - \mu_2; 0, \sigma_1^2 + \sigma_2^2\right)$$

where $ \phi(x; \mu, \sigma^2) $ is the density of a normal distribution with mean $\mu$ and variance $\sigma^2$.

### Solution:

---------------------

#### Conjugate table

We know from the conjugate table that when

$X_i \mid  \theta \sim N (\theta, \sigma^2)$ and  $\theta \sim N(\mu, \tau^2)$, then the posterior is:

 $\theta\mid X\sim N\left(\frac{\tau^2}{\tau^2 +\sigma^2/n}\bar{X} + \frac{\sigma^2/n}{\tau^2 + \sigma^2/n}\mu, \frac{\tau^2\sigma^2/n}{\tau^2+ \sigma^2/n}\right)$

In our case, $\bar{X} = 1$ and $n=1$.

Using this fact, we know that the first part of our posterior mixture, from the likelihood $N(\mu, 1)$ and prior $N(-1, 1)$, will be:
$$
\begin{align}
&N\left(\frac{1}{1 + 1/1}1 + \frac{1/1}{1 + 1/1}(-1), \frac{1(1)/1}{1 + 1/1}\right) \\
&N\left(0, \frac{1}{2}\right)
\end{align}
$$
And the second part, from the likelihood $N(\mu, 1)$ and prior $N(1, 1)$, will be:
$$
\begin{align}
&N\left(\frac{1}{1 + 1/1}1 + \frac{1/1}{1 + 1/1}(1), \frac{1(1)/1}{1 + 1/1}\right) \\
&N\left(1, \frac{1}{2}\right)
\end{align}
$$
What we don't know are the weights of the posterior mixtureâ€”the conjugate table wouldn't help us there.
$$
\mu \mid y \sim (w)N\left(0, \frac{1}{2}\right) + (1-w)N\left(1, \frac{1}{2}\right)
$$
#### Hint 1 equation

We can use the following relationship to try and find the weights and mixture components.

$$\phi(x; \mu_1, \sigma_1^2)\phi(x; \mu_2, \sigma_2^2) = \phi\left(x; \frac{\mu_1/\sigma_1^2 + \mu_2/\sigma_2^2}{1/\sigma_1^2 + 1/\sigma_2^2}, \frac{1}{1/\sigma_1^2 + 1/\sigma_2^2}\right)\phi\left(\mu_1 - \mu_2; 0, \sigma_1^2 + \sigma_2^2\right)$$
$$
\begin{align}
\pi(\mu \mid y) &\propto N(\mu, 1)\left[.5N(-1, 1) + .5N(1, 1) \right] \\
&\propto \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}(y - \mu)^2} \left( \frac{.5}{\sqrt{2\pi}} e^{-\frac{1}{2}(\mu - (-1))^2} + \frac{.5}{\sqrt{2\pi}} e^{-\frac{1}{2}(\mu - 1)^2} \right) \\
&\propto .5\phi(\mu, y, 1) \phi(\mu, -1, 1) + .5\phi(\mu, y, 1) \phi(\mu, 1, 1) && \text{Putting it in the terms of hint 1. In the likelihood, } \mu =\bar{y} = y \text{ in this case.}\\
&\propto.5\phi(\mu, \frac{y/1 + (-1/1)}{1/1 + 1/1}, \frac{1}{1/1 + 1/1}) \phi(y - (-1), 0, 1 + 1) + .5\phi(\mu, \frac{y/1 + (1/1)}{1/1 + 1/1}, \frac{1}{1/1 + 1/1}) \phi(y - 1, 0, 1 + 1)  && \text{Using Hint 1 equation.} \\
&\propto.5\phi(\mu, \frac{y -1}{2}, \frac{1}{2}) \phi(y + 1, 0, 2) + .5\phi(\mu, \frac{y + 1}{2}, \frac{1}{2}) \phi(y - 1, 0, 2)  && \text{} \\
&\propto.5\phi(\mu, 0, \frac{1}{2}) \phi(2, 0, 2) + .5\phi(\mu, 1, \frac{1}{2}) \phi(0, 0, 2)  && \text{Plug in y.} \\
\end{align}
$$
Looking at this, we can see the posterior mixture we identified above. In fact, we could have skipped using the conjugate table and used this directly, but it's useful to check our work and compare to previous class methods.

#### Finding the weights

Let's try to identify $w$ and $1 - w$. 

Does $w = .5 \phi\left(2; 0, 2\right)$ and $(1-w) = .5 \phi\left(2; 0, 2\right)$? Not quite. Let's check that:

````python
from scipy.stats import norm

a = .5 * norm.pdf(2, loc=0, scale=(2)**.5)
b = .5 * norm.pdf(0, loc=0, scale=(2)**.5)

print(a, b)
````

These are not normalized since they don't sum to 1, and therefore our posterior is not "collectively exhaustive" like we learned from the law of total probability. We need to normalize the weights:

````python
w = a/(a + b)
print(w, 1 - w)
````

Then our posterior is $.269N(0, \frac{1}{2}) + .731N(1, \frac{1}{2})$

#### Finding the exact weights (unnecessary!)

What if we didn't have the hint? Could we find the weights? Let's start with Bayes' Theorem, where $\epsilon$ and $(1-\epsilon)$ are the prior mixture weights:
$$
\begin{align}
\pi(\mu \mid y) &= \frac{f(y \mid \mu) \pi(\mu)}{m(y)} \\
&= \frac{f(y \mid \mu)(\epsilon \pi_1(\mu) + (1 - \epsilon)\pi_2(\mu))}{\int f(y \mid \mu)(\epsilon \pi_1(\mu) + (1 - \epsilon)\pi_2(\mu)) d\mu} \\
&= \frac{\epsilon f(y \mid \mu)\pi_1(\mu) + (1 - \epsilon) f(y \mid \mu)\pi_2(\mu)}{\epsilon \int f(y \mid \mu)\pi_1(\mu) d\mu + (1 - \epsilon) \int f(y \mid \mu)\pi_2(\mu) d\mu} \\
\end{align}
$$
Using the following:

$m_i(y) = \int_{-\infty}^{\infty}f(y \mid \mu)\pi_i(\mu)d\mu$,

$ m_1(y) \pi_1(\mu \mid u) = f(u \mid \mu) \pi_1(\mu) $,

and $m_2(y) \pi_2(\mu \mid y) = f(y \mid \mu) \pi_2(\mu)$,
$$
\begin{align}
\pi(\mu \mid y) &= \frac{\epsilon m_1(y) \pi_1(\mu \mid y) + (1 - \epsilon) m_2(y) \pi_2(\mu \mid y)}{\epsilon m_1(y) + (1 - \epsilon) m_2(y)} \\
&= \frac{\epsilon m_1(y)\pi_1(\mu \mid y) + m_2(y) - \epsilon m_2(y)}{\epsilon m_1(y) + m_2(y) - \epsilon m_2(y)} \\
&= \epsilon' \pi_1(\mu \mid y) + (1 - \epsilon')\pi_2(\mu \mid y)
\end{align}
$$
where $\epsilon' = \frac{\epsilon m_1(y)}{\epsilon m_1(y) + (1 - \epsilon) m_2(y)}$. I also gave this result in my last office hours, so you could start here if you went this route.

So now that we've demonstrated the equation, let's calculate our marginals.
$$
\begin{align}
m_1(y) &= \int_{-\infty}^{\infty}f(y \mid \mu)\pi_1(\mu)d\mu \\
&= \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}(y - \mu)^2} \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}(\mu + 1)^2}\\
&= \frac{1}{2\pi}\int_{-\infty}^{\infty} e^{-\frac{1}{2}(1 - \mu)^2 -\frac{1}{2}(\mu + 1)^2} \\ 
&= \frac{1}{2\pi} \int_{-\infty}^{\infty}e^{- \mu^2 - 1} \\
&= \frac{1}{2\pi} \int_{-\infty}^{\infty}e^{-1}e^{- \mu^2} \\
&= \frac{e^{-1}}{2\pi} \int_{-\infty}^{\infty} e^{-\frac{1}{2} \left(\frac{\mu^2}{\frac{1}{2}}\right)} && \text{In the exponent, multiply by } 1=\frac{\frac{1}{2}}{\frac{1}{2}}\\
&= \frac{e^{-1}}{2\pi} \frac{(\sqrt{2\pi(\frac{1}{2})}}{1}\int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi(\frac{1}{2})}}e^{-\frac{1}{2} \left(\frac{\mu^2}{\frac{1}{2}}\right)} && \text{Multiply by } 1=\frac{(\sqrt{2\pi(\frac{1}{2})}}{1}\frac{1}{\sqrt{2\pi(\frac{1}{2})}}  \\
&= \frac{e^{-1}}{2\pi} \frac{\sqrt{\pi}}{1} = \frac{1}{e2\sqrt{\pi}}  && \text{Normal distribution PDF integrates to 1.}\\
\end{align}
$$

$$
\begin{align}
m_2(y) &= \int_{-\infty}^{\infty}f(y \mid \mu)\pi_1(\mu)d\mu \\
&= \int_{-\infty}^{\infty}\frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}(y - \mu)^2} \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}(\mu - 1)^2} d\mu\\
&= \frac{1}{2\pi} \int_{-\infty}^{\infty}e^{-\frac{1}{2}(1 - \mu)^2 -\frac{1}{2}(\mu - 1)^2} d\mu && \text{Using } (1-\mu)^2 = (\mu - 1)^2 \\ 
&= \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{- (\mu - 1)^2} d\mu \\
&= \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{- \frac{\frac{1}{2}(\mu - 1)^2}{\frac{1}{2}}} d\mu && \text{In the exponent, multiply by } 1=\frac{\frac{1}{2}}{\frac{1}{2}} \\
&= \frac{1}{2\pi} \frac{\sqrt{2\pi (\frac{1}{2})}}{1}\int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi(\frac{1}{2})}}e^{- \frac{\frac{1}{2}(\mu - 1)^2}{\frac{1}{2}}} d\mu && \text{Multiply by } 1=\frac{(\sqrt{2\pi(\frac{1}{2})}}{1}\frac{1}{\sqrt{2\pi(\frac{1}{2})}}  \\
&= \frac{1}{2\pi} \frac{\sqrt{2\pi(\frac{1}{2})}}{1}\int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi(\frac{1}{2})}}e^{- \frac{\frac{1}{2}(\mu - 1)^2}{\frac{1}{2}}} d\mu && \text{Normal distribution PDF integrates to 1.}  \\
&= \frac{1}{2\pi} \frac{\sqrt{\pi}}{1} = \frac{1}{2\sqrt{\pi}}  \\
\end{align}
$$

Finally we normalize:
$$
\begin{align}
\epsilon^\prime &= \frac{.5m_1(y)}{.5m_1(y) + .5m_2(y)} \\
\epsilon^\prime &= \frac{.5\frac{1}{e2\sqrt{\pi}}}{.5\frac{1}{e2\sqrt{\pi}} + .5\frac{1}{2\sqrt{\pi}}} \\
\epsilon^\prime &= \frac{\frac{1}{e}}{\frac{1}{e} + 1} = \frac{1}{1+e} \\
\text{and} \\
1- \epsilon^\prime &= 1 - \frac{1}{1+e} = \frac{e}{e + 1} \\
\end{align}
$$
So our exact posterior is:
$$
\frac{1}{1+e}N(0, \frac{1}{2}) + \frac{e}{e + 1}N(1, \frac{1}{2})
$$

## Q3

Our model is:
$$
\begin{align}
y \mid \theta_1, \theta_2 &\sim N(\theta_1 + \theta_2, 1) \\
\theta_i &\sim N(0, \nu^2),\quad i = 1, 2 \\
\nu^2 &\sim \text{Inv. Gamma}(a, b) \\
\end{align}
$$
With one observation of $y=.5$.

First, we derive the full conditionals for all three parameters.
$$
\begin{align}
\pi(\theta_1, \theta_2, \nu^2 \mid y) &\propto f(y \mid \theta_1, \theta_2)\pi(\theta_1 \mid \nu^2)\pi(\theta_2 \mid \nu^2)\pi(\nu^2) \\
&\propto \exp \left( -\frac{1}{2} (y - (\theta_1 + \theta_2))^2 \right) \frac{1}{\sqrt{\nu^2}} \exp \left( -\frac{\theta_1^2}{2\nu^2} \right)\frac{1}{\sqrt{\nu^2}} \exp \left( -\frac{\theta_2^2}{2\nu^2} \right) (\nu^2)^{-a - 1} \exp\left(-b/\nu^2 \right) \\
\end{align}
$$
For $\theta_1$:
$$
\begin{align}
&\propto \exp \left( -\frac{1}{2}(y - (\theta_1 + \theta_2))^2 \right) \exp \left( -\frac{\theta_1^2}{2\nu^2} \right) \\
&\propto \exp \left( -\frac{1}{2}\left(y^2 - 2y\theta_1 - 2y\theta_2 + \theta_1^2 +2\theta_1\theta_2 + \theta_2^2 + \frac{\theta_1^2}{\nu^2} \right)\right)   \\
&\propto \exp \left( -\frac{1}{2}\left(\theta_1^2 - 2y\theta_1 + 2\theta_1\theta_2 + \frac{\theta_1^2}{\nu^2} \right)\right) \\
&\propto \exp \left( -\frac{1}{2}\left(\theta_1^2\left(1 + \frac{1}{\nu^2}\right) - 2\theta_1(y - \theta_2)\right)\right) \\
&\propto \exp \left( -\frac{1}{2}\left(1 + \frac{1}{\nu^2}\right)\left(\theta_1^2 - 2\theta_1\frac{(y - \theta_2)}{\left(1 + \frac{1}{\nu^2}\right)}\right)\right) \\
&\propto \exp \left( -\frac{1}{2}\left(1 + \frac{1}{\nu^2}\right)\left(\theta_1^2 - 2\theta_1\frac{(y - \theta_2)}{\left(1 + \frac{1}{\nu^2}\right)} + \left(\frac{(y - \theta_2)}{\left(1 + \frac{1}{\nu^2}\right)}\right)^2 \right)\right) \\
&\propto \exp \left( -\frac{1}{2}\left(1 + \frac{1}{\nu^2}\right)\left(\theta_1 - \frac{(y - \theta_2)}{\left(1 + \frac{1}{\nu^2}\right)}\right)^2\right) \\
&= N\left(\frac{y - \theta_2}{1 + \frac{1}{\nu^2}}, \frac{1}{1 + \frac{1}{\nu^2}}\right) \text{ or } N\left(\frac{y - \theta_2}{\frac{\nu^2}{\nu^2 + 1}}, \frac{\nu^2}{\nu^2 + 1}\right)
\end{align}
$$

$\theta_2$ will be $N\left(\frac{y - \theta_1}{1 + \frac{1}{\nu^2}}, \frac{1}{1 + \frac{1}{\nu^2}}\right)$ or $N\left(\frac{y - \theta_1}{\frac{\nu^2}{\nu^2 + 1}}, \frac{\nu^2}{\nu^2 + 1}\right)$ by symmetry.

For $\nu^2$:
$$
\begin{align}
&\propto \frac{1}{\sqrt{\nu^2}} \exp \left( -\frac{\theta_1^2}{2\nu^2} \right)\frac{1}{\sqrt{\nu^2}} \exp \left( -\frac{\theta_2^2}{2\nu^2} \right) (\nu^2)^{-a - 1} \exp\left(-b/\nu^2 \right) \\
&\propto \frac{1}{\nu^2} \exp \left( -\frac{\theta_1^2}{2\nu^2}-\frac{\theta_2^2 }{2\nu^2}- \frac{b}{\nu^2} \right)  (\nu^2)^{-a - 1} \\
&\propto (\nu^2)^{-a - 1 - 1} \exp \left( \frac{-.5\theta_1^2- .5\theta_2^2 - b}{\nu^2} \right) \\
&\propto (\nu^2)^{-(a + 1) - 1} \exp \left( \frac{-(.5\theta_1^2+ .5\theta_2^2 + b)}{\nu^2} \right) \\
&= \text{IG}(a + 1, .5(\theta_1^2+ \theta_2^2) + b)
\end{align}
$$

You can also see ```Fall23MidtermQ3code.py``` for the code, but I'll paste it below:

````python
import numpy as np
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style
import pandas as pd
from scipy.stats import invgamma


def stats(df):
    style.use("bmh")
    fig, axs = plt.subplots(ncols=3, figsize=(12, 4))

    for i, col in enumerate(df.columns):
        col_mean = df[col].mean()
        col_credible_interval = np.quantile(df[col], [0.025, 0.975])

        print(f"{col}:")
        print(f"\tMean: {col_mean:.3f}")
        print(
            f"\t95% Equitailed Credible Interval: [{col_credible_interval[0]:.3f}",
            f"{col_credible_interval[1]:.3f}]",
        )

        sns.histplot(df[col], ax=axs[i], stat="density")
        axs[i].set_xlabel(col)
        if i != 0:
            axs[i].set_ylabel("")

    fig.savefig("Q3histograms.png")


def norm_var(v2):
    return 1/(1 + (1/v2))


def norm_loc(other_theta, v2, y):
    return (y - other_theta) * norm_var(v2)


rng = np.random.default_rng(333)

burn = 500
samples = 100000 + burn

theta1_array = np.zeros(samples)
theta2_array = np.zeros(samples)
v2_array = np.zeros(samples)

# params
a = b = 10
y = .5

# inits
v2 = 0.5
theta1 = theta2 = 1


for i in tqdm(range(samples)):
    theta1 = rng.normal(loc=norm_loc(theta2, v2, y), scale=norm_var(v2) ** 0.5)
    theta1_array[i] = theta1

    theta2 = rng.normal(loc=norm_loc(theta1, v2, y), scale=norm_var(v2) ** 0.5)
    theta2_array[i] = theta2

    ig_scale = 0.5 * (theta1**2 + theta2**2) + b
    # v2 = 1 / rng.gamma(shape=a + 1, scale=1 / ig_scale)
    v2 = invgamma.rvs(a=a + 1, scale=ig_scale) # equivalent
    v2_array[i] = v2

v2s = v2_array[burn:]
theta1s = theta1_array[burn:]
theta2s = theta2_array[burn:]

results = pd.DataFrame(
    {"$\\nu_2$": v2_array, "$\\theta_1$": theta1_array, "$\\theta_2$": theta2_array}
)

stats(results)
````
